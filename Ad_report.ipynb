```python
import pandas as pd
```

```python
df = pd.read_csv("data/search_cleaned.csv")
df.head(5)
```

```python
df.loc[df['Campaign Name'].str.contains('Rainforest'), 'Campaign Name'] = "Rainforest"
df.loc[df['Campaign Name'].str.contains('Floral'), 'Campaign Name'] = "Floral"
```

```python
df['Key'] = df['Customer Search Term'] + ' ' + df['Campaign Name']+ ' ' + df['Ad Group Name']+ ' ' + df['Targeting']
```

```python
df['Short Key'] = df['Customer Search Term'] + ' ' + df['Campaign Name']
```

```python
df['Number of Days'].replace({0: 1}, inplace=True)
```

```python
df.columns
```

```python
df.replace('-', 0)
```

```python
df['Impressions Per Day'] = df['Impressions']/df['Number of Days']
df['Spend Per Day'] = df['Spend']/df['Number of Days']
df['Sales Per Day'] = df['7 Day Total Sales ']/df['Number of Days']
df
```

```python
sort_df = df.sort_values(by='Sales Per Day', ascending=False).head(10)
```

```python
sort_df
```

```python
df['Total Advertising Cost of Sales (ACoS) '] = df['Total Advertising Cost of Sales (ACoS) '].fillna(0)
```

```python
#Top search terms to reasess
Zero = df.loc[df['Total Advertising Cost of Sales (ACoS) '] == 0]
Zero
```

```python
zero_df = Zero.sort_values(by='Spend', ascending=False).head(10)
zero_df
```

```python
df.to_csv(r'data\ad2.csv', index = False)
```
